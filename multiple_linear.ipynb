```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
#step1
df=pd.read_csv("car data.csv")
df.head()
```

```python
#step2
df.isna().sum()
```

```python
#step3 data pre-processing
df["Age"]=2026-df["Year"]
df.head()
```

```python
#drop Year columns because dosen't require 
df.drop(columns='Year',inplace=True)
```

```python
df.drop(columns='Car_Name',inplace=True)
```

```python
df
```

```python
df=pd.get_dummies(data=df,drop_first=True)#drop_first actual columns delete kari de 
df
```

```python
y=df["Selling_Price"]
x=df[['Present_Price', 'Kms_Driven', 'Owner', 'Age',
       'Fuel_Type_Diesel', 'Fuel_Type_Petrol', 'Seller_Type_Individual',
       'Transmission_Manual']]
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)#2 dimnesion
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train,y_train)#modal bani gayu
```

```python
m=lr.coef_#  amaa total 8 columns che etla 8 o/p aavyu etla y=m1x1+m2x2+m3x3+m4x4+m5x5+...+m8x8+c
m
```

```python
c=lr.intercept_
c
```

```python
#i20 car /8year old /petrol/manual/45000km/owner 0/8.5lakh/indivisual
i20_price=lr.predict(df[['Present_Price', 'Kms_Driven', 'Owner', 'Age',
       'Fuel_Type_Diesel', 'Fuel_Type_Petrol', 'Seller_Type_Individual',
       'Transmission_Manual']])
```

```python
#i20 car /8year old /petrol/manual/45000km/owner 0/8.5lakh/indivisual
i20_price=lr.predict([[8.5,45000,0,8,0,1,1,1]])#olumns maa jem data hoy tem naakho
i20_price
```

```python
#modeal ketlu accurate ketlo che e check karva ketli error che e shodhva mate 

#aama 3 columns genrate thase x_test(value predict karva mate ni columns) /y_test(actual value of car)/y_pred(model) (car ni value predicat kare)

y_pred=lr.predict(x_test)#predict kari ne aapiyo answer package 
y_pred

from sklearn.metrics import mean_squared_error
print("MSE=",mean_squared_error(y_test,y_pred))
```

```python
import pandas as pd
import numpy as np

df=pd.read_csv("RealEstate.csv")
df.head(5)
```

```python
df.isna().sum()
```

```python
df.drop(columns="No",inplace=True)
```

```python
df=pd.get_dummies(data=df,drop_first=True)#drop_first actual columns delete kari de 
df
```

```python
y=df["Y house price of unit area"]
x=df[['X1 transaction date', 'X2 house age',
       'X3 distance to the nearest MRT station',
       'X4 number of convenience stores', 'X5 latitude', 'X6 longitude',]]
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=110)
print(x_train.shape)#2 dimnesion
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train,y_train)#modal bani gayu
```

```python
m=lr.coef_#  amaa total 8 columns che etla 8 o/p aavyu etla y=m1x1+m2x2+m3x3+m4x4+m5x5+...+m8x8+c
m
```

```python
c=lr.intercept_
c
```

```python
predict=lr.predict(df[['X1 transaction date', 'X2 house age',
       'X3 distance to the nearest MRT station',
       'X4 number of convenience stores', 'X5 latitude', 'X6 longitude',]])
predict
```

```python
y_pred=lr.predict(x_test)#predict kari ne aapiyo answer package 
y_pred

from sklearn.metrics import mean_squared_error
print("MSE=",mean_squared_error(y_test,y_pred))
```

```python

from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_pred,y_test)
print("MAE=",mae)
```

```python
from sklearn.metrics import mean_squared_error
mae=mean_squared_error(y_pred,y_test)
print("MAE=",mae)
```
