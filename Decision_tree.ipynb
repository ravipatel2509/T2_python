# Decision Tree(classification Algorithm)


```python
#step 1 
import pandas as pd
df=pd.read_csv("DecisionTreeDataset.csv")
df.head()
```

```python
#step2
df.isna().sum()
```

```python
#convert into true or false value
#numrical form maa data hoy to ja getdummies use karvu
df=pd.get_dummies(data=df,drop_first=True)
df.head()
```

```python
#step3
y=df["Job Offered_Yes"]
x=df[['CGPA_Low', 'CGPA_Medium', 'Communication_Good', 'Apptitude_Low',
       'Programming Skill_Good']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
model=DecisionTreeClassifier(criterion='entropy',max_depth=5,random_state=42)
model
```

```python
model.fit(x_train,y_train)
model
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
TN=cm[0][0]
FP=cm[0][1]
FN=cm[1][0]
TP=cm[1][1]
print("TN=",TN)
print("FP=",FP)
print("FN=",FN)
print("TP=",TP)
```

```python
Accuracy=(TP+TN)/(TP+TN+FP+FN)
print("Accuracy",Accuracy)
Error=(FP+FN)/(TP+TN+FP+FN)
print("Error:",Error)
Sensitivity=TP/(TP+FN)#TP nu opposite FN
print("Sensitivity",Sensitivity)
Speciticity=TN/(TN+FP)
print("Speciticity:",Speciticity)
precision=TP/TP+FP
print("Precision",precision)
```

```python
import matplotlib.pyplot as plt
plot_tree(model)
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```

```python
#without formula
from sklearn.metrics import accuracy_score,recall_score
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
error=1-accuracy
print(error)
senscitvity=recall_score(y_test,y_pred)
print(senscitvity)
```
