```python
#step1
import pandas as pd
df=pd.read_csv("windpower.csv")
df.head()
```

```python
#step2 
df.isna().sum()
```

```python
#step3
df.dropna(inplace=True)
df.head()
```

```python
y=df["Power (kW)"]
x=df[["Wind speed (m/s)"]]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```
